# Subqueries, Common Table expression and Union

- Writing subqueries
    - Using a subqueries in an IN list
    - Using a subqueries and NOT IN
    - Using subquery containing NULL and NOT IN

```
SELECT <select list> FROM <table1>
WHERE <col1> IN (SELECT <col2> FROM <table2>)
```

````sql
USE AdventureWorks2012

SELECT CustomerID, AccountNumber
from Sales.Customer
WHERE CustomerID IN (select CustomerID from Sales.SalesOrderHeader);
````

````sql
-- Using Subquery and NOT IN

USE AdventureWorks2012

SELECT CustomerID, AccountNumber
FROM sales.Customer
WHERE CustomerID NOT IN 
    (SELECT CustomerID from Sales.SalesOrderHeader);
````

````sql
USE AdventureWorks2012

SELECT CurrencyRateID, FromCurrencyCode, ToCurrencyCode
FROM sales.CurrencyRate
WHERE CurrencyRateID NOT IN
    (SELECT CurrencyRateID from sales.SalesOrderHeader);
````
