# **Joining Tables**

Joining are mainly two type.

- **_INNER JOIN_** : When connecting two tables with _INNER JOIN_, only the rows from the tables that match on the joiing columns will show up in the results.
- **_OUTER JOIN_** :
    - **_LEFT OUTER JOIN_** :
    - **_RIGHT OUTER JOIN_**

```sql
-- INNER JOIN 

USE AdventureWorks2012

SELECT TOP (100)
    soh.SalesOrderID, 
    soh.OrderDate, 
    soh.TotalDue, 
    sod.SalesOrderDetailID,
    sod.ProductID,
    sod.OrderQty
FROM sales.SalesOrderHeader As soh 
INNER JOIN sales.SalesOrderDetail as sod 
    ON soh.SalesOrderID = sod.SalesOrderID;
```

Avoid incorrect joining

```sql
USE AdventureWorks2012

select TOP (100)
    s.SalesOrderID, 
    s.OrderDate, 
    s.TotalDue, 
    d.SalesOrderDetailID,
    d.ProductID,
    d.OrderQty
FROM sales.SalesOrderHeader as s
INNER JOIN Sales.SalesOrderDetail as d 
    ON 1 = 1;
```

### Joining different columns

```sql
USE AdventureWorks2012

SELECT TOP(100)
    c.CustomerID,
    c.PersonID,
    p.BusinessEntityID,
    p.LastName
FROM sales.Customer as c 
INNER JOIN Person.Person as p 
    ON c.PersonID = p.BusinessEntityID;
```

### Joining more than one columns

```sql
USE AdventureWorks2012

SELECT sod.SalesOrderID,
    sod.SalesOrderDetailID,
    so.ProductID,
    so.SpecialOfferID,
    so.ModifiedDate

FROM sales.SalesOrderDetail as sod 
INNER JOIN Sales.SpecialOfferProduct as so 
    ON so.ProductID = sod.ProductID AND
    so.SpecialOfferID = sod.SpecialOfferID
WHERE sod.SalesOrderID IN (51116, 51112)
```
