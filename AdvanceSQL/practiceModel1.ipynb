# **Window Function**
\## ROW\_NUMBER()
### Applying the Row Number to different columns

```sql
USE AdventureWorks2019;
GO

--- Row numbers applied by CustomerID
SELECT CustomerID, 
        SalesOrderID,
        ROW_NUMBER() OVER(ORDER BY CustomerID) as RowNumber
```
