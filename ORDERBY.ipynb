## ORDER BY Clause

```sql
USE AdventureWorks2012

SELECT TOP(20) CustomerID, SUM(TotalDue) AS TotalPerCustomer
FROM Sales.SalesOrderHeader
GROUP BY CustomerID
ORDER BY CustomerID;
```

```sql
USE AdventureWorks2012

SELECT TOP(10) CustomerID, SUM(TotalDue) as TotalPerCustomer
FROM Sales.SalesOrderHeader
GROUP BY CustomerID
ORDER BY MAX(TotalDue) DESC;
```

```sql
USE AdventureWorks2012

SELECT CustomerID, SUM(TotalDue) AS TotalPerCustomer
FROM Sales.SalesOrderHeader
GROUP BY CustomerID
ORDER BY TotalPerCustomer DESC;
```
